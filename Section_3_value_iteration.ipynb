## Implement the Value Iteration algorithm

**Your task.** In the empty cells below, implement *value iteration* yourself, using the
explanation here as your guide.

**The idea in plain language.**
Value iteration is a shortcut compared with policy iteration. Instead of fully evaluating a
policy and then improving it, it folds both into a single sweep: in each pass over the grid you
set every state's value to the value of its *best* action (a `max`, not a probability-weighted average).

For each state:
1. try all 4 actions and compute `Q(s, a) = reward + gamma * V(next_state)` for each;
2. set the state's new value to the largest `Q` (its best action);
3. remember which action was best - that becomes the policy for this state.

Because you always take the best action, the values climb toward the *optimal* values. Keep
sweeping until no value changes by more than a tiny amount (`theta`); then both the values and
the greedy policy they imply are optimal.

**Helpful building blocks:**
- `env.simulate_step(state, action)` returns `(next_state, reward, done, info)`.
- `gamma` (e.g. 0.99) discounts future rewards; `theta` (e.g. 1e-6) is the stopping threshold.
- The `max` over actions is the key difference from policy evaluation, which averages over actions.
