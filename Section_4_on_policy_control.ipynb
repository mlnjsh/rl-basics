**Your task in this notebook: build on-policy Monte Carlo control from scratch.**

### What you are building, in plain language

The agent must learn a good policy for the maze **purely by trial and error**. It keeps a
table `Q(s, a)` ("how good is action `a` in state `s`?") and improves those numbers from
the episodes it plays.

- **On-policy** means the agent **learns about the same policy it uses to act**. Here that
  policy is **epsilon-greedy**: most of the time it takes the best-known action (exploit),
  but with probability `epsilon` it takes a random action (explore) so it keeps
  discovering the maze.
- **Monte Carlo** means it learns from **complete episodes** using the **real returns** it
  observed -- no guessing about the future.

### Algorithm outline (fill in the TODOs below)

For each of many episodes:

1. **Play a full episode** with the epsilon-greedy policy, recording every
   `(state, action, reward)`.
2. **Walk backwards** through the episode, computing the return at each step:
   `G = reward_t + gamma * G` (the discount `gamma` makes sooner rewards count a little
   more).
3. For each `(state, action)` pair, **average all the returns** seen for it and store that
   as the new `Q(s, a)`.

Because the policy reads from `Q`, improving the numbers automatically improves the policy.
