## Implement the algorithm

### What you are building: Q-Learning (off-policy TD control)

Like SARSA, Q-Learning is a **temporal-difference** method - it updates **after every
step** and **bootstraps** (learns from its own guess of the next state's value).
The big new idea is that it is **off-policy**: it uses two different policies.

- A **behaviour / exploratory policy** $b(s)$ decides how the agent actually *moves*
  (above, this is a fully random policy that explores everywhere).
- A **target policy** $\pi(s)$ is the one we are actually *learning about* (above,
  this is a greedy policy - always take the best-known action).

Because we explore with one policy but learn the value of a different greedy one,
the method is **off-policy**.

**The update rule you will implement:**

$$Q(S_t, A_t) \leftarrow Q(S_t, A_t) + \alpha\,\bigl[\,R_{t+1} + \gamma\,\max_{a} Q(S_{t+1}, a) - Q(S_t, A_t)\,\bigr]$$

**SARSA vs Q-Learning - the only real difference:** in the next-state term,

- SARSA uses $Q(S_{t+1}, A_{t+1})$ (the action actually taken next), while
- Q-Learning uses $\max_a Q(S_{t+1}, a)$ (the *best* possible next action).

Since the target policy above is greedy, asking it for the next action is the same as
taking that $\max$.
