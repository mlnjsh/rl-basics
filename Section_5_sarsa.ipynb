### What you are building: SARSA (a temporal-difference method)

So far Monte Carlo methods waited until the **end of an episode** to learn.
**Temporal-difference (TD)** learning, used by SARSA, instead updates the Q-table
**after every single step** - it does not wait. To do this it **bootstraps**: it
improves its estimate of one state using its *current guess* of the next state.

**Why the name SARSA?** Each update needs five things in order:

$$S_t,\; A_t,\; R_{t+1},\; S_{t+1},\; A_{t+1}$$

state, action, reward, next state, and the **next action chosen by the same policy**.

**The update rule you will implement:**

$$Q(S_t, A_t) \leftarrow Q(S_t, A_t) + \alpha\,\bigl[\,R_{t+1} + \gamma\,Q(S_{t+1}, A_{t+1}) - Q(S_t, A_t)\,\bigr]$$

- $\alpha$ = learning rate (how big each step is)
- $\gamma$ = discount factor (how much future reward counts)
- the bracket is the **TD error** (new estimate minus old estimate)

SARSA is **on-policy**: it learns about the very policy it uses to explore (an
epsilon-greedy policy), because it updates using $A_{t+1}$, the action it will
*actually take* next.
