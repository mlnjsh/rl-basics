**The idea in plain language.**
Policy iteration repeats two steps until the plan stops changing:

1. **Policy evaluation** - *How good is my current plan?*
   Keeping the policy fixed, compute a value `V(s)` for every state: the expected total
   (discounted) future reward you would collect starting there and following the current policy.
   You do this by sweeping every state repeatedly and updating its value from its neighbours,
   until the values barely change.

2. **Policy improvement** - *Can I do better?*
   Using those values, in every state pick the action that leads to the best-valued outcome.
   This makes the policy *greedy* with respect to the current values.

Evaluate, then improve, then evaluate the improved policy, and so on. Each round the policy
gets at least as good as before. When an improvement step changes *nothing*, the policy is
optimal and you stop.

**Helpful building blocks (you do not have to use them all):**
- `env.simulate_step(state, action)` returns `(next_state, reward, done, info)` for trying an

- `theta` (e.g. 1e-6) is a tiny threshold: stop evaluating once no value changes by more than it.
- The action-value `Q(s, a) = reward + gamma * V(next_state)` tells you how good an action is.
