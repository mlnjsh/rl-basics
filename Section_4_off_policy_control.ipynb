## Define the target policy $\pi(s)$

Off-policy learning uses **two policies at once**:

- the **target policy** $\pi(s)$ -- the policy you actually want to perfect. It is
  **purely greedy** (always the best-known action, no exploration). This is the policy you
  will test at the end.
- the **behaviour policy** $b(s)$ (defined further down) -- the policy that actually moves
  the agent around during training. It explores.

The whole point of off-policy methods: **explore with $b$, but learn about $\pi$.**

## Implement the algorithm

**Your task: build off-policy Monte Carlo control with importance sampling.**

### On-policy vs off-policy

- **On-policy** (earlier notebooks): the agent learns about the **same** policy it acts
  with, so it can only learn a slightly-random policy.
- **Off-policy** (here): the agent **acts** with the exploratory behaviour policy $b$ but
  **learns** the values of the fully greedy target policy $\pi$. This lets it explore
  freely while still learning the *optimal* (non-random) policy.

### The catch and the fix

The episodes come from $b$, but we want returns as if they came from $\pi$. We correct for
this mismatch with **importance sampling**: each step carries a weight `W` answering *"how
much more likely was this action under $\pi$ than under $b$?"*

Walking **backwards** through the episode:

1. Update the return `G = reward_t + gamma * G`.
2. Update `Q(s, a)` toward `G` using **weighted importance sampling** (track a running
   total weight per state-action pair, and move `Q` by `(W / total_weight) * (G - Q)`).
3. If the behaviour action **differs** from what the greedy target would do, the target
   would never go this way -- so **stop** and start the next episode.
4. Otherwise **grow `W`** by `1 / b(action | state)` and continue.
